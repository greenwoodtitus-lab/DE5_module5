```python
import pandas as pd
import numpy as np
```

```python
customers = pd.read_csv("data/03_Library SystemCustomers.csv")
customers = customers.dropna()
customers['Customer ID'] = customers['Customer ID'].astype(int)
customers.to_csv('data/cleaned_customers.csv', index=False)
```

```python
books = pd.read_csv("data/03_Library Systembook.csv")
books = books.dropna(subset=['Id'])
books['Books'] = books['Books'].str.strip().str.title()
books['Book checkout'] = books['Book checkout'].str.replace('"','',regex=False)
books['Id'] = books['Id'].astype(int)
books['Days allowed to borrow'] = np.where(books['Days allowed to borrow'] == "2 weeks", "14", "MISSING")
books['Days allowed to borrow'] = books['Days allowed to borrow'].astype(int)
books['Book checkout'] = pd.to_datetime(books['Book checkout'], dayfirst=True, errors='coerce')
books['Book Returned'] = pd.to_datetime(books['Book Returned'], dayfirst=True, errors='coerce')

errors = books[books.isnull().any(axis=1)]
errors.to_csv('data/errors.csv', index=False)

books = books.dropna()
books['Customer ID'] = books['Customer ID'].astype(int)
books.to_csv('data/cleaned_books.csv', index=False)
```
